```python
#Importing libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.random.randint(0,20,3)
print(x)
```

```python
xnames= ['group_a','group_b','group_c']
```

```python
# Creating the size of the figure
plt.figure(figsize=(9, 3))

#First subplot is a bar, figure 1
plt.subplot(131)
plt.bar(xnames, x)

#Second subplot is a line, figure 2
plt.subplot(132)
plt.plot(xnames, x)
plt.suptitle('Bar, Line, & Scatter Plotting')

#Third subplot is a scatter, figure 3
plt.subplot(133)
plt.scatter(xnames, x)

plt.show()
```

```python
#Creating a pie chart with clockwise explode slice from tutorial based off % of time spent on course work
#Figure 4

labels = 'Supplemental Videos', 'Assignments', 'Reading', 'Quizes'
sizes = [15, 30, 50, 5]  #Must equal 100%
explode = (0, 0.1, 0, 0)  # "explode" the 2nd slice with is 0.01 (i.e. 'Assignments')

fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%', #Added % at 1 dec point
        shadow=True, startangle=90)
ax1.axis('equal')  # From tutorial: Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```
