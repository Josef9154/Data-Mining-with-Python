1. Create an array of size [10,50] and use the random function to fill it with integers between 1 and 500 (inclusive). Load it to a DataFrame.

2. Calculate the sum of each row; the sum of each column; and the sum of all entries.

3. Determine the minimum; maximum; average; and standard deviation of all entries.

4. Sort the DataFrame: by rows, by columns, on row 2, on column 2.

```python
# Loading Libraries
import pandas as pd
import random
import numpy as np
```

```python
#Setting Decimal point preference
np.set_printoptions(precision =2, suppress =True)
```

```python
#Creating an array using randint as mentioned. 
arr = np.random.randint(1,500,(10,50)) # random numpy array of shape (4,5)
print("Random Array with 1-50 integers with a shape of 10,50:") 
print ()
print (arr)
```

```python
#Load to a dataframe
df = pd.DataFrame (arr)
df
```

```python
#Calculating sum of each row and column and entries. 
 

print("\nSum of array : ", np.sum(arr))    ## Sum of total array
print()
print("Sum of array(axis = 0) : ", np.sum(arr, axis = 0)) ## Sum of columns
print()
print("Sum of arrary(axis = 1) : ", np.sum(arr, axis = 1))  ##Sum of rows
  
print("\nSum of array (keepdimension is True): \n", 
      np.sum(arr, axis = 1, keepdims = True)) 
```

```python
#Descriptive statistics of Array

print("Min of array : ", np.min(arr)) 
print("Max of array : ", np.max(arr)) 
print("Mean of array : ", np.mean(arr)) 
print("Standard Deviation of array : ", np.std(arr))  
```

```python
#Sorting the array by column 
x =arr[arr[:, 1].argsort()] # Sorted by column
x
```
